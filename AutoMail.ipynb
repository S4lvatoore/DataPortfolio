```python
smtp_server = 'smtp.ex.com'         # Адрес SMTP сервера (например, smtp.gmail.com)
smtp_port = 465                          # Порт (обычно 587 для TLS)
sender_email = 'Sended User Mail'  # Ваш email
receiver_email = 'RESEIVER MAIL'  # Email получателя
password = 'Password'      #Пароль отправителя
```

```python
def send_notification(Items, Quantity, Status):
    message = MIMEMultipart()
    message['From'] = sender_email
    message['To'] = receiver_email
    message['Subject'] = f'Уведомление: {Items} - {Status}'
    
    body = f"Продукт: {Items}\nКоличество: {Quantity}\nСтатус: {Status}"
    message.attach(MIMEText(body, 'plain'))
    
    try:
        # Для Mail.ru чаще используют SMTP_SSL
        server = smtplib.SMTP_SSL(smtp_server, smtp_port)
        server.login(sender_email, password)
        server.sendmail(sender_email, receiver_email, message.as_string())
        server.quit()
        print(f"Уведомление отправлено для {Items}")
    except Exception as e:
        print(f"Ошибка при отправке уведомления для {Items}: {e}")
```

```python
csv_file_path = 'PATH_TO_FILE'
```

```python
df = pd.read_csv(csv_file_path, encoding='cp1251')
```

```python
for index, row in df.iterrows():
    product_name = row['Items']
    quantity = row['Quantity']
    
    # Логика определения статуса
    if quantity < 3:
        status = "Out of Stock" if quantity == 0 else "Running Low"
        # Отправляем уведомление
        send_notification(product_name, quantity, status)
    else:
        print(f"{product_name} - In Stock")
```
